```python
import os
import codecs
import spacy
import numpy
import gensim
from spacy import displacy
import pandas as pd
```

```python
ocrd_corpus = 'ALL_DE'
corpus = []
directory = "../corpus_samples/OCRd/" + ocrd_corpus
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if '.txt' in filename: 
        path = os.path.join(directory, filename)
        file = codecs.open(path, encoding="utf8")
        lines = file.readlines()
        stripped_lines = []
        for line in lines:
            stripped_lines.append(line.rstrip())
            separator = ' '
            text = separator.join(stripped_lines)
        text_stripped = text.rstrip()
        corpus.append(text_stripped)
        continue
    else:
        continue
corpus
```

```python
my_stop_words = [u'|',u'||', u'herr', u'mit',u'der',u'die',u'das', u'in',u'sein',u'dies',u'einen',u'und', u'ich', u'mit',u'zu', u'von',u'bis', u'S.', u'°', u'den', u'dem', u'v.',u'u.',u'Z.',u'o.',u'U.',u'A.']
```

```python
%%time
clean_corpus = []
print ('loading german model')
nlp = spacy.load("de_core_news_md")
```

```python
%%time

def discard_token(token):
    if(token.is_space or token.is_stop or token.is_punct 
        or token.like_num or token.is_space or (token.__len__() < 3)):
        return True
    else:
        return False

for stopword in my_stop_words:
    lexeme = nlp.vocab[stopword]
    lexeme.is_stop = True
for page in corpus:
    doc = nlp(page)
    tokens = [token.lemma_ for token in doc if not discard_token(token)]
    print(tokens)
    clean_corpus.append(' '.join(tokens))
```

```python
clean_corpus[0]
```

```python
from gensim.models import Phrases
```

```python
for text in clean_corpus:
    token_array = text.split(' ')
    phrases = Phrases(token_array, min_count=30, progress_per=10000)
phrases
```

```python
with codecs.open(os.path.join(directory +'/processed/', 'de' + '.txt'), 'w', encoding='utf_8') as language_txt_file:
    language_txt_file.write('\r'.join(clean_corpus))
```
